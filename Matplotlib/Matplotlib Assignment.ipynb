### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data  plot a scatter plot. 

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3) 
x = 3 + np.random.normal(0, 2, 50) 
y = 3 + np.random.normal(0, 2, len(x)) 
colors = 3 + np.random.normal(0, 2, len(x)) 
plt.scatter(x,y,s=100,c=colors,cmap='summer')
plt.colorbar().set_label('Grading')
plt.xlabel('X-axis')
plt.xlabel('y-axis')
plt.tight_layout()
plt.show()
```

### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data: 


```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500]) 
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70]) 
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60]) 
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a single figure for all subplots
fig, _ = plt.subplots(nrows=2, ncols=2)

# Create separate variables for the axes
ax1 = fig.axes[0]
ax2 = fig.axes[1]
ax3 = fig.axes[2]
ax4 = fig.axes[3]

# Plot data in each subplot
ax1.plot(x1, y1)
ax1.set_title('Plot 1')

ax2.plot(x2, y2)
ax2.set_title('Plot 2')

ax3.plot(x3, y3)
ax3.set_title('Plot 3')

ax4.plot(x4, y4)
ax4.set_title('Plot 4')

# Adjust layout (optional)
plt.tight_layout()

# Show the single figure with all subplots
plt.show()
```

### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

```python

import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np 
company = np.array(["Apple", "Microsoft", "Google", "AMD"]) 
profit = np.array([3000, 8000, 1000, 10000])

fig,(ax1,ax2)=plt.subplots(nrows=2,ncols=1)


ax1.bar(company,profit,width=0.6)
ax1.set_xlabel('companies')
ax1.set_ylabel('profits in (USD)')
ax2.barh(company,profit,height=0.6)
ax2.set_xlabel('profits in (USD)')
ax2.set_ylabel('companies')

plt.tight_layout()
plt.show()
```

### Q5: What is a box plot? Why is it used? Using the following data plot a box plot. 

```python
import numpy as np
import matplotlib.pyplot as plt
box1 = np.random.normal(100, 10, 200) 
box2 = np.random.normal(90, 20, 200) 

boxes=[box1,box2]
plt.boxplot(boxes)
plt.show()
```
