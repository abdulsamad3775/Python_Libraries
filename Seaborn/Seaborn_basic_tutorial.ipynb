## SEABORN

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.
Seaborn uses fewer syntax and has stunning default themes


```python
# pip install seaborn
```

```python
#importing required library
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#import dataset
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.relplot(x='total_bill',y='tip',data=tips)
```

## Hue, Style and Size

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='smoker')
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker', style = 'time')
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips, hue = 'smoker', style = 'time', size= 'size')
```

## Line plot

```python
sns.set(font_scale=1)
sns.relplot(x='total_bill',y='tip',data=tips,kind='line',height=5,aspect=4,label='tip')
plt.legend(loc='right')
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,kind='line',height=5,aspect=4,hue='smoker')
```

## Subplot

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='smoker',col='smoker')
```

```python
sns.set(font_scale=1)
sns.relplot(x='total_bill',y='tip',data=tips,hue='smoker',col='size')
```

```python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker', col = 'time', data = tips, row = 'size')
```

```python
sns.relplot(x = 'total_bill', y = 'tip', hue = 'smoker', col = 'size', data = tips, col_wrap=3)
```

## Categorical Data Ploting

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips,hue='day')
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips,hue='day', jitter = False)
```

```python
sns.catplot(x = 'day', y = 'total_bill', data = tips, kind = 'swarm',hue='day')
```

```python
sns.catplot(x = 'smoker', y = 'tip', data = tips, order = ['No', 'Yes'],hue='smoker')
```

# Box plot

```python
sns.catplot(x = 'day', y = 'total_bill', kind = 'box', data = tips, hue = 'sex')
```

# Bar plot

```python
sns.catplot(x = 'day', y = 'total_bill', kind = 'bar', data = tips, hue = 'sex')
```

# Point plot

```python
sns.catplot(x = 'day', y = 'total_bill', kind = 'point', data = tips, hue = 'smoker')
```

# Joint Plot

```python
sns.jointplot(x='total_bill', y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip',data=tips,kind='hex')
```

# Pair plot

```python
sns.pairplot(tips)
```

# Regression Plot 


```python
sns.regplot(x = 'total_bill', y = 'tip', data = tips)
```
